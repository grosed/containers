```r
linked_list_node <- function()
    {
      e <- new.env(hash=FALSE,parent=emptyenv())
      e$previous_node <- e$next_node <- e$data <- NULL
      #class(e) <- "linked_list_node"
      return(e)
    }
```

```r
linked_list <- function()
    {
        e <- new.env(hash=FALSE,parent=emptyenv())
        e$front <- e$back <- NULL
        e$size <- 0L
        #class(e) <- "linked_list"
        return(e)
    }
```

```r
push_back <- function(l,data)
    {
        node <- linked_list_node()
        node$data <- data
        if(is.null(l$front) & is.null(l$back)) # list is empty
           {
               l$front <- l$back <- node
           }
        else # non empty list
            {
               node$previous_node <- l$back
               l$back$next_node <- node
               l$back <- node
            }
        l$size <- l$size + 1L
    }
```

```r
pop_front <- function(l)
    {
       if(is.null(l$front))
           {
              return(NULL)
           }
       data <- l$front$data
       l$front <- l$front$next_node
       if(!is.null(l$front))
           {
             l$front$previous_node <- NULL
           }
    l$size <- l$size - 1L
    return(data)
    }
```

```r
l <- linked_list()
n <- 5000000
for(i in 1:n)
    {
    push_back(l,i)
    }
print(l$size)
```

```r
linked_list_node <- setRefClass("linked_list_node", fields = c("pred","succ","data"))
linked_list <- setRefClass("linked_list", fields = c("front","back","size"))
```

```r
push_back <- function(l,data)
    {
        node <- linked_list_node$new(pred=NULL,succ=NULL,data=data)
        if(is.null(l$front) & is.null(l$back)) # list is empty
           {
               l$front <- l$back <- node
           }
        else # non empty list
            {
               node$pred <- l$back
               node$pred$succ <- node
               l$back <- node
            }
        l$size <- l$size + 1L
    }
```

```r
linked_list_node <- function()
    {
      # first entry is predecessor
      # second entry is successor
      # third entry is payload
      return(list(NULL,NULL,NULL))
    }
```

```r
linked_list <- function()
    {
        e <- new.env(hash=FALSE)
        e$front <- e$back <- NULL
        e$size <- 0L
        class(e) <- "linked_list"
        return(e)
    }
```

```r
push_back <- function(l,data)
    {
        node <- linked_list_node()
        #print(address(node))
        node[[3]] <- data
        if(is.null(l$front) & is.null(l$back)) # list is empty
           {
               l$front <- l$back <- node
           }
        else # non empty list
            {
               node[[1]] <- l$back
               l$back[[2]] <- node
               l$back <- node
            }
        #a <- l$front
        #b <- l$back
       # print(address(a))
        #print(address(b))
        l$size <- l$size + 1L
    }
```

```r
pop_front <- function(l)
    {
       if(is.null(l$front))
           {
              return(NULL)
           }
       data <- l$front[[3]]
       l$front <- l$front[[2]]
       if(!is.null(l$front))
           {
             l$front[[1]] <- NULL
           }
    l$size <- l$size - 1L
    return(data)
    }
```
